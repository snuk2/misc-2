## Electric Potential



In Lab 1, you were introduced to the concept of electric fields and electric potentials. In this lab, you will be exploring them from the perspective of energy, and how you can use equipotential lines to predict the behavior of a charged particle.

Potential energy is a rather abstract concept, and it is often defined as the energy of an object due to its position in space. You can think of the potential energy of an object as how much kinetic energy the object could potentially have, barring any loss to frictional forces. An object moving from higher to lower potential energy will gain kinetic energy in the process, and thus move faster. Note also the green field arrow near the center of the figure above, making a 90 degrees angle with the equipotential lines. The direction of the field arrow tells us the direction of the force on the object, and thus the direction of the acceleration. *Will an object always move in the direction of the field arrows?*
